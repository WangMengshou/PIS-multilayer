```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import os,sys
sys.path.append(os.path.dirname(os.path.dirname(os.getcwd())))
from methods.utils import government_feedback, media_hospital_feedback
```

```python
# Generate x values
xlm, xmh, gl, gm, gh = 1/3, 2/3, 0.1, 0.5, 0.9

x_values = torch.linspace(0, 1, 50)
response = government_feedback(x_values, xlm, xmh, gl, gm, gh).numpy()
x_values = x_values.numpy()

colors = ['#EE82EE','#4169E1','#2E8B57','#808000','#A0522D','#CD5C5C','#B22222','#696969','#FF8C00']
markers = ['o','s','*','v']

fig, ax = plt.subplots(figsize=(10, 6))
# Plot the functions
plt.figure(figsize=(10, 6))
ax.scatter(x_values[x_values < xlm], response[x_values < xlm], 
            s=100, c=colors[2], marker='o', alpha=0.8)
ax.scatter(x_values[(xmh > x_values)& (x_values > xlm)], response[(xmh > x_values)& (x_values > xlm)],
            s=100, c=colors[1], marker='^', alpha=0.8)
ax.scatter(x_values[x_values > xmh], response[x_values > xmh],
            s=100, c=colors[0], marker='v', alpha=0.8)
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)
ax.set_xlabel(r'$obs\ (N_{EI}/N)$', fontsize=20)
ax.set_ylabel(r'government attention $g$', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)

ax.annotate('', xy=(xlm, gm), xytext=(xlm, gl),
            arrowprops=dict(facecolor=colors[2], edgecolor='none', shrink=0.15))
ax.annotate('', xy=(xmh, gh), xytext=(xmh, gm),
            arrowprops=dict(facecolor=colors[1], edgecolor='none', shrink=0.15))

ax.text(xlm+0.01, (gm + gl)/2-0.01, r'$x_{lm}$', fontsize=20, color=colors[2])
ax.text(xmh+0.01, (gm + gh)/2-0.01, r'$x_{mh}$', fontsize=20, color=colors[1])

ax.text(xlm+0.02, gl-0.02, r'$g_{l}$', fontsize=20, color=colors[2])
ax.text(xmh+0.02, gm-0.02, r'$g_{m}$', fontsize=20, color=colors[1])
ax.text(1+0.02, gh-0.02, r'$g_{h}$', fontsize=20, color=colors[0])

ax.text(xlm - 0.3, gl + 0.05, '$Low\ intensity$', fontsize=20, color=colors[2])
ax.text(xmh - 0.4, gm + 0.05, '$Medium\ intensity$', fontsize=20, color=colors[1])
ax.text(1 - 0.3, gh + 0.05, '$High\ intensity$', fontsize=20, color=colors[0])

plt.show()
fig.savefig(f'../figs/Schematic/f_obs_g.svg', bbox_inches='tight')
```
